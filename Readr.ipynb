Year
%Y (4 digits).
%y (2 digits; 00-69 → 2000-2069, 70-99 → 1970-1999).

Month
%m (2 digits).
%b (abbreviated name, like “Jan”).
%B (full name, “January”).

Day
%d (2 digits).
%e (optional leading space).

Time
%H (0-23 hour format).
%I (0-12, must be used with %p).
%p (a.m./p.m. indicator).
%M (minutes).
%S (integer seconds).
%OS (real seconds).

```r
parse_date("01/02/15", "%m/%d/%y")
```

```r
parse_date("01/Jan/2001", "%d/%b/%Y")
```
